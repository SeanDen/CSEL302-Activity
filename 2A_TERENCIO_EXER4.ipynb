**1. Import Libraries**

```python
import pandas as pd
import numpy as np #numerical operation
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

**2. Creating a DataFrame and Saving to CSV**

```python
data = pd.read_csv('/content/Lesson_4_student_grades_dataset.csv')
data.head()
```

**3. Visualizing the Data**

```python
plt.scatter(data['Hours_of_Study'],data['Grades'], color='cyan')
plt.xlabel('Independent Variable',)
plt.ylabel('Dependent Variable')
plt.title('Scatter Plot of Data')
plt.show()
```

**4. Preparing Data for Modeling**

```python
X = data[['Hours_of_Study']]
Y = data['Grades']
```

**5. Fitting the Linear regression Model**

```python
model = LinearRegression()
model.fit(X,Y)
```

**6. Getting the Coefficients**

```python
slope = model.coef_[0]
intercept = model.intercept_
print("slope:", slope)
print("intercept:", intercept)
```

**7. Visualizig the LinearRegression Line**

```python
plt.scatter(X,Y, color='yellow')
plt.plot(X,model.predict(X), color='orange')
plt.xlabel('Independent Variable')
plt.ylabel('Dependent Variable')
plt.title('Linear Regression')
plt.show()
```

**8. Making Predictions**

```python
new_X = [[np.random.randint(100,200)]]
predicted_Y = model.predict(new_X)
print("Predicted Y for new X:", predicted_Y)
```

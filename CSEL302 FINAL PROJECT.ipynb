2. Exploring Relationships with Boxplots

*Analysis of Math Scores by Parental Level of Education*

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='parental level of education', y='math score', data=df)
plt.xticks(rotation=45)
plt.title('Math Scores by Parental Level of Education')
plt.show()
```

**1. Higher Parental Education Correlates with Higher Math Scores:**

Students whose parents have higher educational qualifications, such as a master's degree or a bachelor's degree, tend to have higher median math scores. This suggests that higher parental education positively influences students' math performance.

**2. Variability in Scores:**

The spread of scores is narrower for students whose parents have higher education levels (e.g., master's degree), indicating more consistent performance in math. Conversely, there is greater variability in scores among students whose parents have lower educational qualifications, such as a high school diploma or some high school education.

**3. Outliers and Lower Scores:**

There are more outliers (students with significantly lower scores) among those whose parents have lower educational levels (high school or some college). This indicates that some students in these categories face significant challenges in math performance.

**4. Median Scores:**

The median math scores show a clear trend where higher parental education levels correspond to higher median scores. For instance, students with parents holding a master's degree have the highest median math scores, while those with parents having some high school education have the lowest median scores.








*Analysis of Reading Scores by Parental Level of Education*

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='parental level of education', y='reading score', data=df)
plt.xticks(rotation=45)
plt.title('Reading Scores by Parental Level of Education')
plt.show()
```

**1. Higher Parental Education and Reading Scores:**

**3. Outliers:**

The presence of outliers is more pronounced among students whose parents have lower educational qualifications, suggesting that some students in these categories struggle more with reading.

*Analysis of Writing Scores by Parental Level of Education*

**3. Median and Outliers:**

The median writing scores are higher for students with parents holding advanced degrees. More outliers are observed among students with parents having lower educational qualifications, highlighting the challenges faced by these students.

*Key Insights from the Correlation Analysis*

**1. Inter-correlation Among Scores:**

There is a strong positive correlation between the math score, reading score, and writing score. Specifically, the correlation between reading and writing scores is almost perfect (0.95), indicating that students who perform well in one of these areas tend to perform well in the others as well. The math score also shows a strong positive correlation with both reading (0.82) and writing scores (0.80).

**2. Gender Differences:**

The correlation between gender (male) and math score is 0.17, suggesting that male students tend to have slightly higher math scores. However, the correlations with reading (-0.24) and writing scores (-0.29) indicate that female students tend to perform better in these subjects.

**3. Race/Ethnicity:**

The correlation values for different race/ethnicity groups show some variation in performance, but these correlations are generally low, suggesting that race/ethnicity alone does not have a strong direct impact on scores.

**4. Parental Education Level:**

Parental education level shows some interesting correlations with student performance. For example, there is a positive correlation between having parents with a master's degree and math scores (0.14), reading scores (0.11), and writing scores (0.13). Conversely, having parents with a high school education shows a slight negative correlation with math scores (-0.13).

**5. Lunch Type:**

There is a noticeable correlation between lunch type and performance. Students with standard lunch (as opposed to free/reduced lunch) tend to score higher in math (0.35), reading (0.28), and writing (0.31). This suggests that students from higher socioeconomic backgrounds (who are more likely to have standard lunch) perform better academically.


**6. Test Preparation Course:**

Completing a test preparation course shows a positive correlation with higher scores across math (0.25), reading (0.18), and writing (0.21). This indicates that such courses can effectively improve student performance.

*SUMMARY*



The analysis reveals several key factors that influence student performance. Gender differences are evident, with male students performing better in math and female students excelling in reading and writing. Socioeconomic status, as indicated by lunch type, plays a significant role, with students having standard lunch performing better overall. Parental education level also positively impacts student scores, particularly when parents have higher educational attainment. Additionally, participation in test preparation courses is associated with better performance, highlighting the value of such programs.

These insights can help educators and policymakers identify areas for targeted interventions to improve student outcomes, such as providing additional support for students from lower socioeconomic backgrounds and encouraging participation in test preparation courses.
